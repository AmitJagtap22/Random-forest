##Performing EDA
Data can be manipulated by 2 things:
1. Null values
2. Duplicates

We need to treat these values before performing any analysis or else they may affect accuracy of prediction

- Null values can either be replaced or removed
- Duplicates can be dropped


```python
df.isnull().sum()
```

'TotalCharges' column contains float number however, the datatype of that column is object.

```python
df['TotalCharges'].value_counts()
```

From above we can observe that there are 11 empty values which are stored as object data type which is resulting in incorrect assignment of data type to the rest of numerical column.

These empty values are not counted in null values

```python
df[df['TotalCharges']==' ']
```

Above 11 records have no values in the TotalCharges column. We can replace these values with mean or NaN

Here as there are only 11 such records which is very few relative to total records, we can directly drop these records

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
#this will convert all empty values to NaN(by coerce) and then change dtype to numeric(to_num)
#replacing and converting to float is done by above single command
```

```python
df.isnull().sum()
```

Now 11 records are detected while checking for null

```python
df.dtypes
```

Above list confirms that the data type of Total Charges column is converted to float and updated in list successfully

```python
df=df.dropna()
```

```python
df.isnull().sum().sum()
```

All null values are removed

```python
df.duplicated().sum()
```

##Encoding with help of Label Encoding

```python
col_list=[]
for i in df.columns:
    if (df[i].dtype=='object')& (i!='Churn'):
        col_list.append(i)

col_list
```

Above 'col_list' contains the list of all object type columns excluding the target column Churn. Label encoding can be directly performed on this col_list

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in col_list:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

All the object columns are converted to numeric datatype after performing label encoding

##Preparing data for applying Random Forest
Dividing the dataset into dependant and independant columns

```python
x=df.iloc[:,:-1] #features
y=df['Churn']    #target
```

```python
x.head()
```

```python
y.head()
```

Now divide the dataset into training and test set

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)    #usual split is 75,25. Train set should always be greater than test data
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier1=RandomForestClassifier(n_estimators=100)    #n_estimators is number of decision trees
classifier1.fit(x_train,y_train)
```

```python
y_pred=classifier1.predict(x_test)
y_pred  #prediction values given by model
#acutal values are stored in y_test
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier1=RandomForestClassifier(n_estimators=200)
classifier1.fit(x_train,y_train)

y_pred=classifier1.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier1=RandomForestClassifier(n_estimators=300)
classifier1.fit(x_train,y_train)

y_pred=classifier1.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

Accuracy is improving generally with increase in n-estimators(no. of decision trees)

```python
n=[100,200,300,400,500,600,700,800,900,1000]

acc=[]
for i in n:
    model=RandomForestClassifier(n_estimators=i)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    a=accuracy_score(pred,y_test)
    acc.append(a)
```

```python
acc
```

```python
d={'Number of trees':n, 'Accuracy':acc}
acc_data=pd.DataFrame(d)
acc_data
```

From above dataframe we can observe that **600** trees will give the highest accuracy of **79%**.

##Conclusion
With this accuracy we can say that close to 80%(79.1%) of predictions gone through this random forest will be correct for newly added customers
